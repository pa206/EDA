```python
# importing the requried libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset
data=pd.read_csv(r"C:\Users\Fundemics\Downloads\auto-mpg.csv")
data
```

```python
'''1.what is the distribution of MPG in the dataset'''
h=data.mpg.values.tolist()
fig=plt.figure(figsize=(6,5),facecolor='green')
axes=plt.axes(facecolor='orange')
sns.histplot(h,color='red',kde=True,label='mpg')
plt.xlabel('mpg')
plt.title('mpg vs count ')
plt.legend()
plt.show()
```

```python
'''2.how many cars are from each origin '''
plt.figure(figsize=(5,5),facecolor='red')
sns.countplot(x=data['origin'],hue=data['origin'])
plt.show()
```

```python
'''3.what is the proportion of cars from each origin '''
plt.figure(figsize=(4,3))

plt.pie(x=data.origin.value_counts().tolist(),colors=['red','blue','green'],labels=data.origin.value_counts().index.tolist()
       ,autopct='%1.2f%%',explode=[0.1,0,0],shadow=True)
plt.legend()
plt.title('Cars')
plt.show()
```

```python
'''4.what is the proportion of cars from each cylinder'''
plt.figure(figsize=(10,3),facecolor='white')
plt.axes(facecolor='orange')
plt.pie(data.cylinders.value_counts().tolist(),labels=data.cylinders.value_counts().index.tolist(),autopct='%1.2f%%',startangle=40,shadow=True,colors=['palegreen','cyan','magenta','yellow','navy'])
plt.title('Cylinders')
plt.show()
```

```python
'''5.'''
plt.figure(figsize=(5,4))
data['horsepower']=pd.to_numeric(data['horsepower'],errors='coerce')
data[['horsepower','weight']].corr()
sns.scatterplot(x=data['horsepower'],y=data['weight'],color='red')
plt.show()
```

```python
'''6.How does MPG vary across different numbers of cylinders?(box)'''
sns.boxplot(data=data,x='cylinders',y='mpg',hue='cylinders')
plt.show()
```

```python
'''7.How does horsepower vary across different numbers of cylinders?(box)'''
sns.boxplot(data=data,x='cylinders',y='horsepower')
plt.show()
```

```python
'''8.What are the relationships among horsepower,and acceleration?  (scatter)'''
sns.scatterplot(data=data,x='horsepower',y='acceleration',color='orange')
plt.show()
```

```python
'''9.How does horsepower vary by origin?  (boxplot)'''
plt.figure(figsize=(4,3))
sns.boxplot(data=data,x='origin',y='horsepower',hue='origin',medianprops=dict(color='red'),capprops=dict(linestyle='--',color='navy'))
plt.show()
```

```python
'''10.What is trend of MPG across different model years?(using line plot)'''
year_mpg = data.groupby('model year')['mpg'].mean()
plt.figure(figsize=(4,3),facecolor='violet')
plt.axes(facecolor='yellow')
plt.plot(year_mpg.index, year_mpg.values,color='cyan',marker='o',markerfacecolor='black',markeredgecolor='navy')
plt.xlabel('model year',color='red',size=15)
plt.ylabel('mpg',color='red',size=15)
plt.title('model year vs mpg',color='black',size=15)
plt.xticks(color='green')
plt.yticks(color='green')
plt.show()
```

```python
'''11.What is the distribution of displacement in the dataset?(using histogram) '''
sns.histplot(data=data,x='displacement',kde=True,)
plt.show()
```
